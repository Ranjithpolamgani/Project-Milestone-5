## Milestone 4: Connecting to an API/Pulling in the Data and Cleaning/Formatting

```python
import requests
import pandas as pd

# Fetch data from the API
url = "https://restfulcountries.com/api/v1/countries"
access_token = "1600|zhNkaJfmn3mqrgE9nhJoZwlbBF7zgK3bLc3A6z8S"
# passing access token through header
headers = {"Authorization": f"Bearer {access_token}"}
# adding headers along with url while performing GET request
response = requests.get(url, headers=headers)
data = response.json()['data']  # Get the data part of the JSON response
# converting json to dataframe using 'json_normalize' method
df = pd.json_normalize(data)
df
```

## Step 1: Replacing Headers

```python
# Change column names to be more readable
# replacing '.'  with '_' in column names
df.columns = [col.replace('.', '_').lower() for col in df.columns]
# 'rename' method is used to rename the columns when a dictionary is passed with key value pairs of current column name and new column name
df.rename(columns={
    'name': 'country_name',
    'full_name': 'full_country_name',
    'capital': 'capital_city',
    'iso2': 'iso_alpha2',
    'iso3': 'iso_alpha3',
    'population': 'population_count',
    'size': 'area_in_sq_km',
    'covid19_total_case': 'total_cases',  
    'covid19_total_deaths': 'total_deaths', 
    'current_president.name': 'current_president',
    'current_president.gender': 'president_gender',
}, inplace=True)  # inplace is used to replace the columns instead of creating new columns
df.head()
```

## Step 2: Identify and drop duplicates

```python
# Checking for duplicate countries based on their ISO codes, country name and removing them using 'drop_duplicates' method
df.drop_duplicates(subset=['iso_alpha2', 'iso_alpha3','country_name'], inplace=True)
df.head()
```

## Step 3: Fix casing

```python
# Converting the string values to title case
df['country_name'] = df['country_name'].str.title()
df['continent'] = df['continent'].str.title()
df['capital_city'] = df['capital_city'].str.title()
# Converting columns to uppercase
df['iso_alpha2'] = df['iso_alpha2'].str.upper()
df['iso_alpha3'] = df['iso_alpha3'].str.upper()
df['currency'] = df['currency'].str.upper()
df.head()
```

## Step 4: Datatype conversions

```python
# We are assigning datatypes to necessary columns so that they can be easily inserted into any databases
df['country_name'] = df['country_name'].astype(str)  # Ensure country names are strings
df['full_country_name'] = df['full_country_name'].astype(str)
df['capital_city'] = df['capital_city'].astype(str)
df['continent'] = df['capital_city'].astype(str)
df['population_count'] = pd.to_numeric(df['population_count'], errors='coerce')
df['total_cases'] = pd.to_numeric(df['total_cases'], errors='coerce')
df['total_deaths'] = pd.to_numeric(df['total_deaths'], errors='coerce')
df['current_president'] = df['current_president'].astype(str)  
df['current_president_appointment_start_date'] = pd.to_datetime(df['current_president_appointment_start_date'], errors='coerce')
df['current_president_appointment_end_date'] = pd.to_datetime(df['current_president_appointment_end_date'], errors='coerce')
df['independence_date'] = pd.to_datetime(df['independence_date'], errors='coerce')
df
```

## Step 5: Translating unicode text

```python
# importing unidecode library
from unidecode import unidecode

# replacing empty values with 'None'
df['full_country_name'].fillna('None', inplace=True)

# applying unidecode to the column
df['full_country_name'] = df['full_country_name'].apply(unidecode)
df
```

## Step 6: Replacing values

```python
import numpy as np

# Replace empty and 'nan' strings in the entire DataFrame with NaN
df.replace('', np.nan, inplace=True)
df.replace('nan', np.nan, inplace=True)

# Removing 'km2' and commas from the 'area_in_sq_km' column and converting the column to numeric
df['area_in_sq_km'] = pd.to_numeric(df['area_in_sq_km'].str.replace(' km²', '').str.replace(',', ''), errors='coerce')
df
```
